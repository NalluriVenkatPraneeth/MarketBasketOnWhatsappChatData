```python
import pandas as pd
```

```python
data = pd.read_csv('wtsapdata.csv')
data = data.iloc[:,1:]
```

```python
data = data.drop(['time','Am/Pm','Message'],axis=1)
data
```

```python
#mlxtend package for association rule algo
#apriori algo and association rules generation
from mlxtend.frequent_patterns import apriori, association_rules

#preprocessing module for transaction encoder
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Grouping the data by user
#Converting that to a list
#using the tolist() to convert the Names to a list which is stored in the variable a as 'Names'
transactions = [a[1]['Name'].tolist() for a in list(data.groupby(["date"]))]

#Initializing the object for TransactionEncoder()
transEnco = TransactionEncoder()

#Using this object to transform this data of 'Names' to a transaction data of binary values
trans_array = transEnco.fit(transactions).transform(transactions)

#converting this array of binary values into a dataframe with column names
transactions1 = pd.DataFrame(trans_array, columns=transEnco.columns_)
```

```python
frequent_itemsets = apriori(df=transactions1,min_support=0.2,use_colnames=True)
```

```python
frequent_itemsets
```

```python
#Generating the association rules from mlxtend package
#imposing min confidence threshold as 50 percentage
rules = association_rules(frequent_itemsets)
```

```python
#Displaying sorted rules in descending order of lift value
ne = rules.sort_values(by='lift',ascending=False)
```
